#### MERGE CON CONTACTS

##### Mergeamos las tablas de contactos y tareas por el id de contacto para tener en un mismos dataframe los datos de ambos ficheros. Seleccionamos las columnas que consideramos apropiadas para llevar a cabo con ellas un posterior análisis

```python
# Inner join
df_tasks_cons = pd.merge(df_con, df_tasks, left_on='con_id', right_on='whoid', how='inner')

# Select specific columns
columns_to_select = ['task_id',
    'msf_seniority__c', 'msf_birthyear__c', 'msf_entrycampaign__c', 'npo02__averageamount__c',
    'msf_begindatemsf__c', 'msf_datefirstrecurringdonorquota__c', 'gender__c', 'msf_ltvcont__c',
    'msf_ltvdesc__c', 'msf_recencyrecurringdonorcont__c', 'msf_rfvrecurringdonor__c', 'npo02__totaloppamount__c',
    'msf_valuetotalcont__c', 'msf_valuetotaldesc__c', 'msf_lifetime__c', 'msf_pressurecomplaint__c', 
    'msf_scoringrfvrecurringdonor__c', 'msf_averagedonorvalue__c', 'msf_percomssummary__c',
    'whoid', 'activitydate', 'msf_channel__c', 'msf_inboundoutbound__c', 'msf_closetype__c'
]
df_tasks_cons = df_tasks_cons[columns_to_select]
```

```python
df_tasks_cons.shape
```

##### Creamos una columna nueva en la que computamos los días que han pasado respecto a la tarea anterior basándonos en la columna activitydate de la tabla de tareas. Lo hacemos así porque pensamos que es interesante disponer de una columna que nos indique los días transcurridos desde la última interacción con el socio.

```python
# Sort by 'whoid' and 'activitydate' to mimic the Window's partitionBy and orderBy in PySpark
df_tasks_cons = df_tasks_cons.sort_values(by=['whoid', 'activitydate'])

df_tasks_cons['previous_modification_date'] = df_tasks_cons.groupby('whoid')['activitydate'].shift(1)

# Calculating the difference in days between the current and previous dates
df_tasks_cons['days_elapsed_since_previous_modification'] = (df_tasks_cons['activitydate'] - df_tasks_cons['previous_modification_date']).dt.days
```

##### Hacemos el merge entre las tablas de donaciones periódicas, la de modificaciones de cuotas y la de contactos (para asociar el id del socio) Seleccionamos las columnas que consideramos apropiadas para el posterior análisis

```python
# Assuming df_con, df_rec, and df_quo are your DataFrames in pandas

# Joining df_con and df_rec on the condition (df_con.con_id == df_rec.npe03__contact__c)
merged_df = df_con.merge(df_rec, left_on='con_id', right_on='npe03__contact__c', how='inner')

# Further joining the result with df_quo on the condition (df_con.con_id == df_quo.msf_contactid__c)
merged_df = merged_df.merge(df_quo, left_on='con_id', right_on='msf_contactid__c', how='inner')

# Selecting the necessary columns
selected_df = merged_df[['con_id', 'rec_id', 'msf_recurringdonation__c',
                         'msf_annualizedquota__c', 'npe03__total_paid_installments__c', 'npe03__installment_period__c',
                         'msf_changedate__c', 'msf_changetype__c', 'msf_leadsource3__c',
                         'msf_newamount__c', 'msf_newannualizedquota__c', 'msf_newrecurringperiod__c']]

# Applying the where conditions
df_rec_quo = selected_df[
    (selected_df['rec_id'] == selected_df['msf_recurringdonation__c'])
]
```

```python
df_rec_quo.shape
```

##### A continuación utilizamos las funciones lag y datediff para meter en cada registro una columna con la fecha del cambio de cuota anterior y los datos con la cuota del anterior cambio. La justificación es que parece interesante tener esta información en todos los registros para luego rellenar información en campos que se hayan quedado en blanco al hacer el siguiente merge que queremos hacer.

```python
# Sort by 'con_id' and 'msf_changedate__c' to mimic the Window's partitionBy and orderBy in PySpark
df_rec_quo = df_rec_quo.sort_values(by=['con_id', 'msf_changedate__c'])

# Creating new columns mimicking the lag function in PySpark
df_rec_quo['previous_change_date'] = df_rec_quo.groupby('con_id')['msf_changedate__c'].shift(1)
df_rec_quo['msf_oldamount__c'] = df_rec_quo.groupby('con_id')['msf_newamount__c'].shift(1)
df_rec_quo['msf_oldannualizedquota__c'] = df_rec_quo.groupby('con_id')['msf_newannualizedquota__c'].shift(1)
df_rec_quo['msf_oldrecurringperiod__c'] = df_rec_quo.groupby('con_id')['msf_newrecurringperiod__c'].shift(1)

# Calculating the difference in days between the current and previous dates
df_rec_quo['days_elapsed_since_previous_change'] = (df_rec_quo['msf_changedate__c'] - df_rec_quo['previous_change_date']).dt.days
```

```python
df_rec_quo.loc[df_rec_quo["con_id"] == '0033Y00002unQGTQA2']
```

##### Dado que no hay forma de cruzar directamente el fichero de tareas con el de modificaciones de cuota lo que hacemos es crear un left join desde tareas en el que el cruce se hace por id de cliente (por supuesto) y de forma que la diferencia de fechas entre la de la tarea y la siguiente modificación de cuota en la línea temporal sea menor de 60 días. Esto lo hacemos así porque la fechas de una tarea que resulta positiva en la apmpliación de cuota no tiene repercusión en la propia tabla de modificaciones de cuota hasta unos días después, dependiendo de la periodicidad de la cuota. Obviamente para periodicidades superiores a la mensual puede que los registros no se asocien correctamente. Por lo que hemos visto el fichero de Tareas comenzó a rellenarse sobre el año 2014 con posterioridad al de donaciones periódicas. Esto hace que en el siguiente merge se pierdan registros, pero es preferible que sea así porque dichos registros corresponderían únicamente a interacciones positivas (incrementos de cuota) y haría que la muestra final no estuviera balanceada

```python
# First, create temporary columns in df_tasks_cons for the lower and upper limits of the date range
df_tasks_cons['upper_date_limit'] = df_tasks_cons['activitydate'] + pd.Timedelta(days=60)

# Now, merge with df_rec_quo using an inner merge to get only matching rows
matched_df = pd.merge(
    left=df_tasks_cons, 
    right=df_rec_quo, 
    left_on='whoid', 
    right_on='con_id',
    suffixes=('', '_right'),
    how='inner'
)

# Filter based on the date conditions
matched_df = matched_df[
    (matched_df['activitydate'] <= matched_df['msf_changedate__c']) & 
    (matched_df['msf_changedate__c'] <= matched_df['upper_date_limit'])
]
```

```python
matched_df.loc[matched_df["con_id"] == '0033Y00002unQGTQA2']
```

```python
# Drop the temporary columns
matched_df.drop(columns=['upper_date_limit'], inplace=True)

# Now, combine the rows from df_tasks_cons with the matched_df using an outer join 
# to retain all rows from df_tasks_cons and only matched rows from df_rec_quo
df_merge = pd.merge(
    df_tasks_cons,
    matched_df,
    left_on='task_id', 
    right_on='task_id',
    how='left'
)
```

##### Rellenamos determinadas columnas que hayan quedado a null con los datos del registro anterior que las tenga rellenas.

```python
# Sort the dataframe by 'whoid', 'activitydate', and 'previous_modification_date'
df_merge = df_merge.sort_values(by=['whoid', 'activitydate', 'previous_modification_date'])

columns_to_fill = [
    "msf_newamount__c", "msf_newannualizedquota__c", "msf_newrecurringperiod__c",
    "previous_change_date", "msf_oldamount__c", "msf_oldannualizedquota__c", "msf_oldrecurringperiod__c"
]

# Fill NaN values with the previous non-NaN value within the same 'whoid' group
for col in columns_to_fill:
    df_merge[col] = df_merge.groupby('whoid')[col].transform(lambda x: x.fillna(method='ffill'))
```

##### Hacemos lo mismo pero esta vez en sentido inverso temporalmente.

```python
# Sort values in preparation for the bfill operation
df_merge.sort_values(by=['whoid', 'activitydate'], ascending=[True, False], inplace=True)

columns_to_fill = ["previous_change_date", "msf_oldamount__c", "msf_oldannualizedquota__c", "msf_oldrecurringperiod__c"]

# Fill NaN values using backward-fill within each 'whoid' group
for col in columns_to_fill:
    df_merge[col] = df_merge.groupby('whoid')[col].transform(lambda x: x.fillna(method='bfill'))

# If you wish to reset the original order, you can sort by 'whoid' and 'activitydate' again
df_merge.sort_values(by=['whoid', 'activitydate'], inplace=True)
```

##### Creamos una columna que nos indique los días transcurridos desde el último cambio de cuota.

```python
df_merge['days_elapsed_since_previous_quota_change'] = (df_merge['activitydate'] - df_merge['previous_change_date']).dt.days
```

##### Rellenamos también valores de columnas que hayan quedado en blanco para algunos de los campos que nos interesan

```python
df_merge['days_elapsed_since_previous_modification'] = df_merge['days_elapsed_since_previous_modification'].combine_first(df_merge['days_elapsed_since_previous_quota_change'])
df_merge['msf_newamount__c'] = df_merge['msf_newamount__c'].combine_first(df_merge['msf_oldamount__c'])
df_merge['msf_newannualizedquota__c'] = df_merge['msf_newannualizedquota__c'].combine_first(df_merge['msf_oldannualizedquota__c'])
df_merge['msf_newrecurringperiod__c'] = df_merge['msf_newrecurringperiod__c'].combine_first(df_merge['msf_oldrecurringperiod__c'])                         
```

##### Transformamos el año de nacimiento en la edad del socio

```python
from datetime import datetime

df_merge['msf_birthyear__c'] = pd.to_numeric(df_merge['msf_birthyear__c'], errors='coerce').fillna(0).astype(int)

# Convert birth year to age
current_year = datetime.now().year
df_merge['age'] = df_merge['msf_birthyear__c'].apply(lambda x: current_year - x if x != 0 else 0)

# Calculate average age, excluding those with age 0
avg_age = df_merge[df_merge['age'] != 0]['age'].mean()

# Replace 0 age with average age
df_merge['age'] = df_merge['age'].replace(0, int(avg_age))
```
