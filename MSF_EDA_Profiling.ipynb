```python
import pandas as pd
```

```python
campaigns = pd.read_parquet('data/MSF_Campaign.parquet')
campaigns.head(3)
```

```python
from ydata_profiling import ProfileReport 

def gen_profiler_report(df, title):
        original_report = ProfileReport(df, title=title  )
        original_report.to_file(f"./profiling_reports/{title}.html")
```

```python
campaign_profiling = gen_profiler_report(campaigns, "Campaign Profiling Report")
campaign_profiling
```
